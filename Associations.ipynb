# **Task:**
### **Dataset:** Var_3_groceries

# **Importing all necessary libraries**

```python
import warnings
warnings.simplefilter('ignore')
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules,fpgrowth
from mlxtend.preprocessing import TransactionEncoder
import time
```

# **Basic database of transactions**

## **Preparing database of transactions**

```python
df=pd.read_csv('/content/Var_3_groceries.csv',delimiter='\t',header=None,names=['Transaction'])
display(df)
```

```python
df['Transaction']=df['Transaction'].str.split(',')
display(df)
```

```python
te=TransactionEncoder()
te_ary=te.fit(df['Transaction']).transform(df['Transaction'])
df=pd.DataFrame(te_ary,columns=te.columns_)
display(df)
```

## **Apriori**

```python
configs=[]
times1=[]
```

```python
min_support=[0.03,0.05,0.07]
max_len=[None,3,2]
min_conf=[0.3,0.34,0.38]
n_runs=10
```

```python
for supp in min_support:
  for conf in min_conf:
    for len in max_len:
      config=f'min_support={supp},min_conf={conf},max_len={len}'
      print(config)
      elapsed_time=0
      for i in range(n_runs):
        start_time=time.perf_counter()
        frequent_itemsets=apriori(df,min_support=supp,max_len=len,use_colnames=True)
        end_time=time.perf_counter()
        elapsed_time+=end_time-start_time
      avg_elapsed_time=round(elapsed_time/n_runs,3)
      print(f'Average elapsed time:{avg_elapsed_time}')
      display(association_rules(frequent_itemsets, metric="confidence",min_threshold=conf)[['antecedents','consequents','antecedent support','consequent support','support','confidence','lift']].sort_values(by='confidence',ascending=False).reset_index(drop=True))
      configs.append(config)
      times1.append(avg_elapsed_time)
```

```python
frequent_itemsets=apriori(df,min_support=0.03,max_len=None,use_colnames=True)
associations=association_rules(frequent_itemsets, metric="confidence",min_threshold=0.3)[['antecedents','consequents','antecedent support','consequent support','support','confidence','lift']].sort_values(by='confidence',ascending=False).reset_index(drop=True)
items=frozenset(['tropical fruit'])
display(associations[associations['antecedents']==items])
```

```python
configs=[]
times2=[]
```

```python
min_support=[0.03,0.05,0.07]
max_len=[None,3,2]
min_conf=[0.3,0.34,0.38]
n_runs=10
```

## **Comparing results**
